```python
import altair as alt
import pandas as pd

# Load CSV file of none track 5 vs Track 5 song ratings
df = pd.read_csv('Taylor Survey Responses.csv')
```

```python
# Display chart of none track 5 songs
chart1 = alt.Chart(df).mark_bar().encode(
    x=alt.X('None Track 5 Song:N', title='Song Name'), 
    y=alt.Y('Individual Rating', title='Rating Per Song on a scale of 1 - 10'),
    
    tooltip=['None Track 5 Song', 'Individual Rating']
)

chart1.properties(
    title='Rating of None Track 5 Songs',
    width=800,
    height=200).interactive()
```

```python
# Display chart of track 5 songs
chart2 = alt.Chart(df).mark_bar().encode(
    x=alt.X('Average Rating of Track 5 Song:N', title='Song Name'), 
    y=alt.Y('Average Rating', title='Rating Per Song on a scale of 1 - 10'),
    color=alt.Color('Average Rating of Track 5 Song', title='Song Name'),
    tooltip=['Average Rating of Track 5 Song', 'Average Rating'])

chart3 = chart2.properties(
    title='Average Rating of Track 5 Songs',
    width=600,
    height=400).interactive()

chart3
```

```python
# Display chart of average rating of track 5 vs none track 5 songs
chart3 = alt.Chart(df).mark_bar().encode(
    x=alt.X('Song Title:N', title='Song Name'), 
    y=alt.Y('Average Rating of None Track 5', title='Rating Per Song on a scale of 1 - 10'),
    color=alt.Color('Song Title', title='Song Name'),
    tooltip=['Average Rating of None Track 5', 'Song Title'])

chart4 = chart3.properties(
    title='Average Rating of Track 5 vs None Track 5 Songs',
    width=400,
    height=400).interactive()

chart4
```

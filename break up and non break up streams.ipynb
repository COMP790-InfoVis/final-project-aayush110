```python
## this code block is for FIGURE 25 and FIGURE 26

import altair as alt
import pandas as pd

# define data for plot1
df1 = pd.DataFrame({
    "type of song": ["(break up)", "(non break up)"],
    "cumulative number of streams (as of Apr 3, 2023)": [2645263857, 5878364443]
})

# create plot1 with selection
brush1 = alt.selection(type='single', empty='none', encodings=['x'])
plot1 = (
    alt.Chart(df1)
    .mark_bar()
    .encode(
        x=alt.X("type of song", sort="y", title=None),
        y="cumulative number of streams (as of Apr 3, 2023)",
        color=alt.Color("type of song",
                         scale=alt.Scale(
                             domain=df1.sort_values(['cumulative number of streams (as of Apr 3, 2023)'])['type of song'].tolist(),
                             range=['magenta', 'purple'])
                        ),
        opacity=alt.condition(brush1, alt.value(1), alt.value(0.5))
    )
    .add_selection(brush1)
)

# define data for plot2
df2 = pd.DataFrame({
    "Does Taylor Swift only write break-up songs?": ["(Yes)", "(No)"],
    "Number of Responses": [11, 97],
    "type of song": ["(break up)", "(non break up)"]
})

# create selection for plot2
brush2 = alt.selection(type='single', empty='none', encodings=['x'])

# create plot2 with filter
plot2 = (
    alt.Chart(df2)
    .mark_bar()
    .encode(
        x=alt.X("Does Taylor Swift only write break-up songs?", sort="y", title=None),
        y="Number of Responses",
        color=alt.Color("Does Taylor Swift only write break-up songs?",
                         scale=alt.Scale(
                             domain=df2.sort_values(['Number of Responses'])["Does Taylor Swift only write break-up songs?"].tolist(),
                             range=['fuchsia', 'violet'])
                        ),
        opacity=alt.condition(brush1 | brush2, alt.value(1), alt.value(0.5))
    )
    .add_selection(brush2)
)

# display the plots side by side
combined_plots = alt.hconcat(plot1, plot2)
combined_plots
```

```python
## this is the code block for FIGURE 4. Click on each of the bars in the first graph for a corresponding bar to pop up on the second graph!
import altair as alt
import pandas as pd

# define data for plot1
df1 = pd.DataFrame({
    "Is Taylor Swift your favorite artist?": ["Yes", "No"],
    "Number of Responses": [63, 45]
})

# create plot1 with selection
brush1 = alt.selection(type='single', empty='none', fields=['Is Taylor Swift your favorite artist?'])
plot1 = (
    alt.Chart(df1)
    .mark_bar()
    .encode(
        x=alt.X("Is Taylor Swift your favorite artist?", sort="y", title="Is Taylor Swift your favorite artist?"),
        y="Number of Responses",
        color=alt.Color("Is Taylor Swift your favorite artist?",
                         scale=alt.Scale(
                             domain=df1.sort_values(['Number of Responses'])['Is Taylor Swift your favorite artist?'].tolist(),
                             range=['darkslategrey', 'cornflowerblue'])
                        ),
        opacity=alt.condition(alt.datum['Is Taylor Swift your favorite artist?'] == 'Yes', alt.value(1), alt.value(0.5))
    )
    .add_selection(brush1)
)

# define data for plot2
df2 = pd.DataFrame({
    "Do you think she is more successful than other artists?": ["(Yes)", "(No)"],
    "Number of Responses": [101, 6],
    "Is Taylor Swift your favorite artist?": ["Yes", "No"]
})

# create selection for plot2
brush2 = alt.selection(type='single', empty='none', fields=['Do you think she is more successful than other artists?'])

# create plot2 with filter
plot2 = (
    alt.Chart(df2)
    .transform_filter(brush1)
    .mark_bar()
    .encode(
        x=alt.X("Do you think she is more successful than other artists?", sort="y", title="Do you think she is more successful than other artists?"),
        y="Number of Responses",
        color=alt.Color("Do you think she is more successful than other artists?",
                         scale=alt.Scale(
                             domain=df2.sort_values(['Number of Responses'])["Do you think she is more successful than other artists?"].tolist(),
                             range=['lightgreen', 'olive'])
                        ),
        opacity=alt.condition(alt.datum['Do you think she is more successful than other artists?'] == '(Yes)', alt.value(1), alt.value(0.5))
    )
    .add_selection(brush2)
)

# combine the two plots horizontally
combined_plots = alt.hconcat(plot1, plot2, data=df2)

# display the combined plots
combined_plots
```
